```python
# Matching tracksub with designations from NEOSurveyor
def match_desig_trksub(offecliptic_file:str, onecliptic_file:str) -> dict:
    """
    Create a dictionary that matches the trksubs to the designations given by the NEOSurveyor team
    """

    file_list = [offecliptic_file, onecliptic_file]
    #empty dictionary 
    trksub_design = {}
    
    for f in file_list:
        #check ig the file exists
        if not os.path.isfile(f):
            print(f"File {f} not found.")
            return None
        # open the file
        with open(f, 'r', encoding='utf-8') as file:
            lines = file.readlines()
            for line in lines:
        
        # create a dictionary to store the matches 
            sp = lines.split()
            trksub = sp[0]
            desig = sp[1]
            if desig[3] =="0":
                # NEO case
                NEA = 1
            else:
                # MBA case
                NEA = 0
            trksub_desig[trksub] = [desig, NEA]
            
    return trksub_desig
```
